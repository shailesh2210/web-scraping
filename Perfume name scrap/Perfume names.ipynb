```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
perfume_name = []
price = []

for i in range(2 , 53):
    url = f"https://www.fridaycharm.com/collections/branded-perfumes-for-men-ladies?page={str(i)}&sort_by=created-descending"
    r = requests.get(url)
    
    soup = BeautifulSoup(r.text , "lxml")
    name = soup.find_all("div", class_ = "grid-product__title grid-product__title--body")
    for i in name:
        name = i.text
        perfume_name.append(name)
```

```python
df = pd.DataFrame({"Perfume Name": perfume_name})
```
